## 1. Configure the run

Set the CSV path and run mode. Use `fast` for development, then `balanced` or `full` for longer runs.

```python
from ts_advisor.config import TSAdvisorConfig
from ts_advisor.logging import ProgressLogger
from ts_advisor.pipeline import run_training_pipeline

CSV_PATH = "data/raw/4G5G_trr_trs_20000.csv"
config = TSAdvisorConfig(run_mode="fast", model_path="models/ts_advisor_model.joblib")
config
```

## 2. Run training and baseline comparison

Progress logs show stage-level timing, data shapes, feature dimensions, model names, feature sets, validation metrics, and warnings.

```python
logger = ProgressLogger()
outputs = run_training_pipeline(CSV_PATH, config=config, logger=logger)
```

## 3. Review data and feature shapes

```python
print("Raw shape:", outputs["raw"].shape)
print("Prepared shape:", outputs["prepared"].shape)
print("Train / Val / Test:", outputs["train_df"].shape, outputs["val_df"].shape, outputs["test_df"].shape)
print("TF-IDF train matrix:", outputs["matrices"].X_train_text.shape)
print("Full train matrix:", outputs["matrices"].X_train_full.shape)
```

## 4. Baseline result table

Confirm that MultinomialNB is only evaluated with TF-IDF-only features.

```python
cols = ["model", "feature_set", "train_accuracy", "val_accuracy", "macro_f1", "weighted_f1", "time_s", "warnings"]
outputs["results_df"][[c for c in cols if c in outputs["results_df"].columns]]
```

## 5. Try an inference call from the saved bundle
